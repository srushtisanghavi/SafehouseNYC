## Load Data Into Postgres

### Creating Connection to Postgres

```python
import psycopg, os

conn = psycopg.connect(
    host="localhost",
    port='5432',
    dbname="zillow_data",
    user="postgres",
    password="123")

print('Connected to PostgreSQL Database')
```

```python
# conn.rollback()
cur = conn.cursor()
```

### Creating Tables based on Functional Dependencies

```python
# Creating table Property Location
createProperty_Location = """ CREATE TABLE IF NOT EXISTS property_location (
                    zpid FLOAT,
                    state VARCHAR(255),
                    street_address VARCHAR(255),
                    latitude VARCHAR(255),
                    longitude VARCHAR(255),
                    borough VARCHAR(255),
                    zipcode FLOAT,
                    PRIMARY KEY (zpid)
                    )
                """

cur.execute((createProperty_Location),)
conn.commit()
```

```python
# Creating table property details 
createProperty_Details = """ CREATE TABLE IF NOT EXISTS property_details (
                    zpid FLOAT,
                    description TEXT,
                    homeStatus VARCHAR(255),
                    yearBuilt FLOAT,
                    FOREIGN KEY (zpid) REFERENCES property_location
                    )
                """

cur.execute(createProperty_Details)
conn.commit()
```

```python
# Creating table photos
createPhotos = """ CREATE TABLE IF NOT EXISTS photos (
                    zpid FLOAT,
                    photos TEXT,
                    FOREIGN KEY (zpid) REFERENCES property_location
                    )
                """

cur.execute(createPhotos)
conn.commit()
```

```python
# Creating table price_tax_info 
createPrice_tax_info = """ CREATE TABLE IF NOT EXISTS price_tax_info (
                    zpid FLOAT,
                    propertyTaxRate VARCHAR(225),
                    tax_annual_amount FLOAT,
                    tax_assessed_value FLOAT,
                    currency VARCHAR(225),
                    price VARCHAR(225),
                    FOREIGN KEY (zpid) REFERENCES property_location
                    )
                """

cur.execute(createPrice_tax_info)
conn.commit()
```

```python
# Creating table Agent Info
createAgent_Info = """ CREATE TABLE IF NOT EXISTS agent_info (
                    zpid FLOAT,
                    agent_info VARCHAR(255),
                    FOREIGN KEY (zpid) REFERENCES property_location
                    )
                """

cur.execute(createAgent_Info)
conn.commit()
```

```python
# Creating table property features 
createproperty_features = """ CREATE TABLE IF NOT EXISTS property_features (
                    zpid FLOAT,
                    living_area FLOAT,
                    occupant_type VARCHAR(255),
                    heating VARCHAR(225),
                    cooling VARCHAR(225),
                    parking VARCHAR(225),
                    bathrooms_full VARCHAR(225),
                    bathrooms_half VARCHAR(225),
                    bedrooms VARCHAR(225),
                    can_raise_horses VARCHAR(225),
                    furnished VARCHAR(225),
                    garage_spaces VARCHAR(225),
                    additional_parcels VARCHAR(225),
                    additional_garage VARCHAR(225),
                    attached_property VARCHAR(225),
                    carport VARCHAR(225),
                    garage VARCHAR(225),
                    home_warrenty VARCHAR(225),
                    land_lease VARCHAR(225),
                    open_parking VARCHAR(225),
                    pets_allowed VARCHAR(225),
                    rent_control VARCHAR(225),
                    spa VARCHAR(225),
                    view VARCHAR(225),
                    parcel_Number VARCHAR(225),
                    parking_slot VARCHAR(225),
                    stories VARCHAR(225),
                    FOREIGN KEY (zpid) REFERENCES property_location
                    )
                """

cur.execute(createproperty_features)
conn.commit()
```

### Inserting Values from Zillow_df into the tables

```python
#Inserting values in property_location table
for index, row in zillow_df.iterrows():
    zpid = row['zpid']
    state = row['State']
    street_address = row['Street_Address']
    latitude = row['latitude']
    longitude = row['longitude']
    borough = row['Borough']
    zipcode = row['Zipcode']
    cur.execute("INSERT INTO property_location (zpid, state, street_address, latitude, longitude, borough, zipcode) VALUES (%s, %s, %s, %s, %s, %s, %s)", (zpid, state, street_address, latitude, longitude, borough, zipcode))
conn.commit()
```

```python
#Inserting values in property_details table
for index, row in zillow_df.iterrows():
    zpid = row['zpid']
    description = row['description']
    homestatus = row['homeStatus']
    yearbuilt = row['yearBuilt']
    cur.execute("INSERT INTO property_details (zpid, description, homeStatus, yearBuilt) VALUES (%s, %s, %s, %s)", (zpid,description, homestatus, yearbuilt))
conn.commit()
```

```python
#Inserting values in Photos table
for index, row in zillow_df.iterrows():
    zpid = row['zpid']
    photos = row['photos']
    cur.execute("INSERT INTO photos (zpid, photos) VALUES (%s, %s)", (zpid, photos))
conn.commit()
```

```python
#Inserting values in Price_tax_info table
for index, row in zillow_df.iterrows():
    zpid = row['zpid']
    propertytaxrate = row['propertyTaxRate']
    tax_annual_amount = row['Tax_Annual_Amount']
    tax_assessed_value = row['Tax_Assessed_Value']
    currency = row['currency']
    price = row['price']
    cur.execute("INSERT INTO price_tax_info (zpid, propertyTaxRate, tax_annual_amount, tax_assessed_value, currency, price) VALUES (%s, %s, %s, %s, %s, %s)", (zpid, propertytaxrate, tax_annual_amount, tax_assessed_value, currency, price))
conn.commit()
```

```python
#Inserting values in Agent_table table
for index, row in zillow_df.iterrows():
    zpid = row['zpid']
    agent_info = row['Agent_Info']
    cur.execute("INSERT INTO agent_info (zpid, agent_info) VALUES (%s, %s)", (zpid, agent_info))
conn.commit()
```

```python
#Inserting values in property_features table
for index, row in zillow_df.iterrows():
    zpid = row['zpid']
    livingArea = row['livingArea']
    occupant_type = row['Occupant_Type']
    heating = row['Heating']
    cooling = row['Cooling']
    parking = row['Parking']
    bathroom_full = row['Bathrooms_Full']
    bathrooms_half = row['Bathrooms_Half']
    bedrooms = row['Bedrooms']
    can_raise_horses = row['Can_Raise_Horses']
    furnished = row['Furnished']
    garage_spaces = row['Garage_Spaces']
    additional_parcels = row['Additional_Parcels']
    additional_garage = row['Additional_Garage']
    attached_property = row['Attached_Property']
    carport = row['Carport']
    garage = row['Garage']
    home_waarenty = row['Home_Warranty']
    land_lease = row['Land_Lease']
    open_parking = row['Open_Parking']
    pets_allowed = row['Pets_Allowed']
    rent_control = row['Rent_Control']
    spa = row['Spa']
    view = row['View']
    parcel_number = row['Parcel_Number']
    parking_slot = row['Parking_Slot']
    stories = row['stories']


    cur.execute("INSERT INTO property_features (zpid, living_area, occupant_type, heating, cooling, parking, bathrooms_full, bathrooms_half, bedrooms, can_raise_horses, furnished, garage_spaces, additional_parcels, additional_garage, attached_property, carport, garage, home_warrenty, land_lease, open_parking, pets_allowed, rent_control, spa, view, parcel_number, parking_slot, stories) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)", (zpid, livingArea, occupant_type, heating, cooling, parking, bathroom_full, bathrooms_half, bedrooms, can_raise_horses, furnished, garage_spaces, additional_parcels, additional_garage, attached_property, carport, garage, home_waarenty, land_lease, open_parking, pets_allowed, rent_control, spa, view, parcel_number, parking_slot, stories))
conn.commit()
```
